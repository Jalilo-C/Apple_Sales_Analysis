### 2. Data Preparation

```python
# Transform the sale_date type column into datetime
apple_sales_df['sale_date'] = pd.to_datetime(apple_sales_df['sale_date'])
```

```python
# Check if revenue match calculations
(apple_sales_df['revenue_usd'] == (apple_sales_df['units_sold'] * apple_sales_df['discounted_price_usd']).round(2)).value_counts()
```

All the rows has been verified, the revenue calculations are right.

* Handling missing values

```python
# Replacing the NaN values in storage and previous device os columns by 'Unknown'
apple_sales_df['storage'] = apple_sales_df['storage'].fillna('Unknown')
apple_sales_df['previous_device_os'] = apple_sales_df['previous_device_os'].fillna('Unknown')
```

Missing values in customer_rating are preserved as NaN to avoid introducing bias.

### 3. Creating additional columns

* Season Column

```python
# Get the season from the date
def get_season(date):
    month = date.month
    day = date.day

    if (month==3 and day>=21) or (month==4 or month==5) or (month==6 and day<21):
        return 'Spring'
    elif (month==6 and day>=21) or (month==7 or month==8) or (month==9 and day<23):
        return 'Summer'
    elif (month==9 and day>=23) or (month==10 or month==11) or (month==12 and day<21):
        return 'Fall'
    else:
        return 'Winter'
    
# Creating the season column
apple_sales_df['season'] = apple_sales_df['sale_date'].apply(get_season)

# Inserting it in the dataset
apple_sales_df.insert(4, "season", apple_sales_df.pop("season"))
```

* Discount Amount Column

```python
# Creating the discount amount column
apple_sales_df["discount_amount"] = apple_sales_df['unit_price_usd'] - apple_sales_df['discounted_price_usd']

# Inserting it in the dataset
apple_sales_df.insert(15, "discount_amount", apple_sales_df.pop("discount_amount"))
```

* Discount Level column

```python
# Creating the discount level column
apple_sales_df['discount_level'] = pd.cut(
    apple_sales_df['discount_pct'],
    bins=[-0.1, 0, 5, 10, 15],
    labels=['No Discount', 'Low', 'Medium', 'High']
)

# Inserting it in the dataset
apple_sales_df.insert(16, 'discount_level', apple_sales_df.pop('discount_level'))
```

* Year-Month Column

```python
# Creating the month-year column
apple_sales_df['year_month'] = apple_sales_df['sale_date'].dt.to_period('M')

# Inserting it in the table
apple_sales_df.insert(6, 'year_month', apple_sales_df.pop('year_month'))
```
